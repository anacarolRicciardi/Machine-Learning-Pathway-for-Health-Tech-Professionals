Perform t-test (After Transformation) -Independent T-test assuming unequal variance

```python
from scipy import stats
from scipy.stats import ttest_ind

#List of variables to test
variables = ['log_serum_creatinine', 'log_serum_sodium', 'log_creatinine_phosphokinase', 'ejection_fraction', 'age', 'time']

#Dictionary to store t-test results
ttest_results = {}

#Perform t-test for each variable
for var in variables:
    #split the data by Death_Events (1, 0)
    group1= df[df['DEATH_EVENT'] == 1][var]
    group0= df[df['DEATH_EVENT'] == 0][var]

    #Perform Welch's t-test (independent t-test with unequal variance)
    t_stat, p_value = stats.ttest_ind(group1, group0, equal_var=False)

   # Calculate degrees of freedom using Welch-Satterthwaite equation
    n1, n2 = len(group1), len(group0)
    s1, s2 = np.var(group1, ddof=1), np.var(group0, ddof=1)
    
    dof = ((s1 / n1 + s2 / n2) ** 2) / (
        (s1 / n1) ** 2 / (n1 - 1) + (s2 / n2) ** 2 / (n2 - 1)
    )
    
    # Calculate t-critical value (two-tailed test, alpha = 0.05)
    alpha = 0.05
    t_critical = stats.t.ppf(1 - alpha / 2, dof)
    
    # Store results
    ttest_results[var] = {
        't-statistic': t_stat,
        'p-value': p_value,
        't-critical (alpha=0.05)': t_critical,
        'degrees of freedom': dof
    }

# Display results
for var, res in ttest_results.items():
    print(f"{var}:")
    print(f"  t-statistic  = {res['t-statistic']:.4f}")
    print(f"  p-value      = {res['p-value']:.4e}")
    print(f"  t-critical   = {res['t-critical (alpha=0.05)']:.4f}")
    print(f"  degrees of freedom = {res['degrees of freedom']:.2f}\n")
```

We take the absolute value of t-statistic and compare it with t-critical value. And, then, we look for p-value if it is below 0.05 in order to test if there is significant difference between the two means. Since for the five variables have t-stat above t-critical, and p-value below 0.05, we conclude that serum_creatinine, serum_sodium, ejection_fraction, follow-up time and age have significant association with Death Events. We neglect creatinine_phosphokinase, since the t-stat is below the t-critical which renders it insignificant

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt

#Step 1: Prepare the Data
x = df[['serum_creatinine', 'serum_sodium', 'ejection_fraction', 'time', 'age']]
y = df[['DEATH_EVENT']]

#Step 2: Split the Data into Train and Test Sets (80/20 split)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#step3: Initialize and fit the logistic regression model
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)

#Step4: make predictions
y_pred = model.predict(x_test)

#step5: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

# Print Results
print(f"Accuracy: {accuracy:.2f}\n")
print("Classification Report:\n", report)

# Step 6: Visualize the Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Step 7: Display Model Coefficients
coefficients = pd.DataFrame({
    'Variable': x.columns,
    'Coefficient': model.coef_[0]
})
print("\nLogistic Regression Coefficients:\n", coefficients)
```

```python

# Step 7: ROC Curve
# Ensure x1_test contains the same features used during training
x1_test = x1_test[['serum_creatinine', 'ejection_fraction', 'time']]

# Generate predicted probabilities for the positive class
y_pred_prob = model.predict_proba(x1_test)[:, 1]  

# Calculate FPR, TPR, and thresholds
fpr, tpr, thresholds = roc_curve(y1_test, y_pred_prob)

# Calculate the AUC
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', label='Random Guess')  # Diagonal line
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.legend(loc='lower right')
plt.grid(alpha=0.2)
plt.show()

# Step 8: Optimal Threshold
optimal_idx = (tpr - fpr).argmax()  # Maximize TPR - FPR
optimal_threshold = thresholds[optimal_idx]
print(f"Optimal Threshold: {optimal_threshold:.2f}")

# Step 9: Display Model Coefficients
coefficients = pd.DataFrame({
    'Variable': x1.columns,
    'Coefficient': model.coef_[0]
})
print("\nLogistic Regression Coefficients:\n", coefficients)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

columns_to_cluster = ['serum_creatinine', 'serum_sodium', 'ejection_fraction', 'time', 'age']

#step 1: extract relevant features
data = df[columns_to_cluster]

#step 2: standardize the features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

#step 3: perform k-means clustering (determine the optimal number of clusters)
inertia = []
k_range = range(1,10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_) #Record interia (sum of squared distances)

# Step 4: Plot the Elbow Curve to find the optimal number of clusters
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.grid()
plt.show()

#step 5 : apply k-means with optimal clusters(k=2)
kmeans = KMeans(n_clusters =2, random_state=42)
df['cluster'] =kmeans.fit_predict(data_scaled)

#step6: visualize clusters 
sns.pairplot(df, hue='cluster', vars=columns_to_cluster, palette='Set2')
plt.title('Cluster Analysis Pair Plot')
plt.show()

#Step 7 : Display Cluster Centers
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
cluster_df = pd.DataFrame(cluster_centers, columns=columns_to_cluster)
print("Cluster Centers:")
print(cluster_df)
```

Calculating the accuracy for this model:

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from scipy.stats import mode
# Assuming `true_labels` contains the ground truth class labels
true_labels = df[['DEATH_EVENT']]
# and `df['cluster']` contains the predicted cluster labels

# Step 1: Map cluster labels to true labels
# Find the mode of the true labels for each cluster
labels_mapping = {}
for cluster_id in np.unique(df['cluster']):
    mask= df['cluster'] == cluster_id
    labels_mapping[cluster_id] = mode(true_labels[mask])[0][0]

# Step 2: Create a new array with predicted labels mapped to true labels
mapped_predictions = df['cluster'].map(labels_mapping)

#step 3: calculate accuracy
accuracy = accuracy_score(true_labels, mapped_predictions)
print(f"Accuracy of clustering model: {accuracy:.2f}")


# Step 4: Optional - Classification Report
print("\nClassification Report:")
print(classification_report(true_labels, mapped_predictions))
```

```python
from sklearn.metrics import cohen_kappa_score

# Extract true labels and predicted clusters from the second clustering model
true_labels = df['DEATH_EVENT']
predicted_clusters = df['cluster1']

# Map cluster labels to match true labels
labels_mapping = {}
for cluster_id in np.unique(predicted_clusters):
    mask = predicted_clusters == cluster_id
    cluster_mode = mode(true_labels[mask]).mode
    labels_mapping[cluster_id] = cluster_mode[0] if isinstance(cluster_mode, np.ndarray) else cluster_mode

# Create predicted labels mapped to true labels
mapped_predictions = predicted_clusters.map(labels_mapping)

# Calculate Cohen's Kappa
kappa = cohen_kappa_score(true_labels, mapped_predictions)

# Print the result
print(f"Cohen's Kappa for the second clustering model: {kappa:.2f}")
```
